```python
def ninetieth_percentile_rows(df, column_name):
    percentile_90 = df[column_name].quantile(0.90)
    return df[df[column_name] >= percentile_90]

ninetieth_percentile = ninetieth_percentile_rows(key_stats, 'R')
ninetieth_percentile.sort_values(by='R', ascending=False)
```

```python
def eightieth_percentile_rows(df, column_name):
    percentile_90 = df[column_name].quantile(0.90)
    percentile_80 = df[column_name].quantile(0.80)
    return df[(df[column_name] >= percentile_80) & (df[column_name]< percentile_90)]

eightieth_percentile = eightieth_percentile_rows(key_stats, 'R')
eightieth_percentile.sort_values(by='R', ascending=False)
```

```python
runs_list_by_decade = []
x = 76
y = 85
while x >= 6:
    runs_mean_by_decade = stats_modern.loc[x:y, 'R'].mean()
    runs_list_by_decade.append(round(runs_mean_by_decade, 2)) 
    x = x - 10
    y = y - 10

runs_list_by_decade.append(round(stats_modern.loc[0:5, 'R'].mean(), 2))

data = {'Decade': ['1940s', '1950s', '1960s', '1970s', '1980s', '1990s','2000s','2010s', '2020s'],
        'Runs': runs_list_by_decade}
df_runs_by_decade = pd.DataFrame(data)
df_runs_by_decade
```

```python
home_runs = stats_modern['HR'].tolist()
runs = stats_modern['R'].tolist()
x = np.array(home_runs)
y = np.array(runs)

mean_x = np.mean(x)
mean_y = np.mean(y)

sd_x = np.std(x, ddof=1)
sd_y = np.std(y, ddof=1)

correlation = np.corrcoef(x, y)[0, 1]
slope = (correlation * sd_y) / sd_x
intercept = mean_y - slope * mean_x
print(f"Slope (b1): {slope}")
print(f"Intercept (b0): {intercept}")

y_pred = intercept + slope * x

print(f"Predicted values: {y_pred}")
```
